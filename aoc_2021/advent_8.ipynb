```python
def output_calculus(code):
  if dict_a in code and dict_b in code and dict_c in code and dict_d not in code and dict_e in code and dict_f in code and dict_g in code:
    return 0
  elif len(code) == 2:
    return 1
  elif dict_a in code and dict_b not in code and dict_c in code and dict_d in code and dict_e in code and dict_f not in code and dict_g in code:
    return 2
  elif dict_a in code and dict_b not in code and dict_c in code and dict_d in code and dict_e not in code and dict_f in code and dict_g in code:
    return 3
  elif len(code)== 4:
    return 4
  elif dict_a in code and dict_b in code and dict_c not in code and dict_d in code and dict_e not in code and dict_f in code and dict_g in code:
    return 5
  elif dict_a in code and dict_b in code and dict_c not in code and dict_d in code and dict_e in code and dict_f in code and dict_g in code:
    return 6
  elif len(code) == 3:
    return 7
  elif len(code) == 7:
    return 8
  else:
    return 9
```

```python
total = 0
for i in range(n):
  temp = data[0][i].split('|')
  output = temp[1].split()
  input = temp[0].split()
  idx_1 =  np.where([len(x)==2 for x in input])[0][0]
  idx_7 =  np.where([len(x)==3 for x in input])[0][0]
  idx_4 =  np.where([len(x)==4 for x in input])[0][0]
  idx_8 =  np.where([len(x)==7 for x in input])[0][0]

  dict_a = list(set(input[idx_7]) - set(input[idx_1]))[0]
  set_4_7 = set(input[idx_4]).union(set(input[idx_7]))
  temp1 = [set(x)-set_4_7 for x in input]

  dict_g = next(iter(temp1[np.where([len(x) == 1 for x in temp1])[0][0]]))
  temp2 = [x - set(dict_g) for x in temp1]

  dict_e = next(iter(temp2[np.where([len(x) == 1 for x in temp2])[0][0]]))
  set_7_g = set(input[idx_7]).union(dict_g)
  temp3 = [set(x)-set_7_g for x in input]

  dict_d = next(iter(temp3[np.where([len(x) == 1 for x in temp3])[0][0]]))
  temp4 = [set(x)-set([dict_a,dict_d,dict_g,dict_e]) for x in input]

  dict_c = next(iter(temp4[np.where([len(x) == 1 for x in temp4])[0][0]]))

  temp5 = set(input[idx_1])-set(dict_c)
  dict_f = next(iter(temp5))

  temp6 = set(input[idx_8])-set([dict_a,dict_c,dict_d,dict_e,dict_f,dict_g])
  dict_b = next(iter(temp6))

  out_coded = [output_calculus(x) for x in output]
  total = total + out_coded[0]*1000 + out_coded[1] * 100 + out_coded[2]*10 + out_coded[3]
total
```
