```python
import pandas as pd
import numpy as np
data = pd.read_table('sample_data/input_7.txt',header = None)
```

```python
crab_p_ch = np.array(data[0][0].split(','))
crab_position = [int(p) for p in crab_p_ch]
```

```python
test_p = np.arange(min(crab_position),max(crab_position)+1)

dist = []
for p in test_p:
  dist.append(sum([abs(p - i) for i in crab_position]))




```

```python
def crab_fuel(n):
  return n * (n + 1) * 0.5
```
