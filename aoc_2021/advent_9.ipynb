```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_table('sample_data/input_9.txt',header = None)
n = data.shape[0]
m = n
```

```python
# matrix with test input
data_t = []

i = 0
for i in range(n):
  data_t.append([int(j) for j in data[0][i].split(',')])

dat_mat = np.asmatrix(data_t)
```

```python
# matrix with input
data_t = []
for i in range(n):
  data_t.append([int(j) for j in data[0][i]])

dat_mat = np.asmatrix(data_t)
```

```python
total = 0
low_points = []
n = dat_mat.shape[0]
m = dat_mat.shape[1]
for i in range(n):
  for j in range(m):
    cond = True
    if i > 0 :
      cond = cond and dat_mat[i,j] < dat_mat[i - 1,j]
    if i < (n-1):
      cond = cond and dat_mat[i,j] < dat_mat[i + 1,j] 
    if j > 0:
      cond = cond and dat_mat[i,j] < dat_mat[i,j - 1]
    if j < (m-1):
      cond = cond and dat_mat[i,j] < dat_mat[i,j + 1]
    
    if cond:
      
      low_points.append([i,j])
      total = total + 1 + dat_mat[i,j]

total
```

```python
def basin(x,y,n,m):
  neighbors = []
  if [x,y] not in neighbors:
    neighbors.append([x,y])
  if x > 0:
    if dat_mat[x,y] < dat_mat[x - 1,y] and dat_mat[x - 1,y] < 9 and [x-1,y] not in neighbors:
      neighbors.extend(basin(x-1,y,n,m))
  if x < (n-1):
    if dat_mat[x,y] < dat_mat[x + 1,y] and dat_mat[x + 1,y] < 9 and [x+1,y] not in neighbors:
      neighbors.extend(basin(x+1,y,n,m))
  if y > 0:
    if dat_mat[x,y] < dat_mat[x,y - 1 ] and dat_mat[x,y - 1] < 9 and [x,y - 1] not in neighbors:
      neighbors.extend(basin(x,y-1,n,m))
  if y < (m-1):
    if dat_mat[x,y] < dat_mat[x,y + 1] and dat_mat[x,y + 1] < 9 and [x,y + 1] not in neighbors:
      neighbors.extend(basin(x,y+1,n,m))
  return neighbors
```

```python
# part 2 

len_basin = []

for j in range(len(low_points)):
  [x,y] = low_points[j]
  test_list = basin(x,y,n,m)
  res = []
  for i in test_list:
    if i not in res:
        res.append(i)
  len_basin.append(len(res))
  
len_basin
```

```python
res = 1
for i in range(3):
  tmax = max(len_basin)
  print(tmax)
  len_basin.remove(tmax)
  res = res * tmax
res
```
