```python
import pandas as  pd
import numpy as np
data = pd.read_table('sample_data/input_5.txt',header = None)
n = 1000
m = data.shape[0]
```

```python
# advent 5 part 1

ocean_floor = np.zeros((n,n))

for i in range(m):
  coords = data[0][i].split(' -> ')
  [x1,y1] = coords[0].split(',')
  [x2,y2] = coords[1].split(',')
  x1 = (int(x1))
  y1 = (int(y1))
  x2 = (int(x2))
  y2 = (int(y2))
  if x1 == x2:
    for j in np.arange(min(y1,y2),max(y1,y2) + 1):
      ocean_floor[x1][j] = ocean_floor[x1][j]+1
  if y1 == y2:
    for j in np.arange(min(x1,x2),max(x1,x2) + 1):
      ocean_floor[j][y1] = ocean_floor[j][y1]+1
print(ocean_floor)
sum(sum(ocean_floor > 1))
```
