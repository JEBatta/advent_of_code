```python
import pandas as pd
from math import floor
import numpy as np
from scipy import stats

data = pd.read_table('sample_data/input_3.txt',header = None)

matrix = np.zeros((len(data),12))
```

```python
for j in range(1000):  
  n = str(data[0][j])
  for i in range(len(n)):
    matrix[j][i] = int(n[len(n)-1-i])
```

```python
gamma = 0
eps = 0
for k in range(12):
  g_bit = stats.mode(matrix[:,k])[0][0]
  if g_bit == 1:
    e_bit = 0
  else:
    e_bit = 1
  gamma = gamma + g_bit * 2**k
  eps = eps + e_bit * 2**k
gamma * eps
```

```python
def my_mode(array,empate):
  n_ones = sum(array)
  n_zeros = sum(array == 0)
  if n_ones > n_zeros:
    return 1
  elif n_ones < n_zeros:
    return 0
  else:
    return empate

def my_antimode(array,empate):
  n_ones = sum(array)
  n_zeros = sum(array == 0)
  if n_ones > n_zeros:
    return 0
  elif n_ones < n_zeros:
    return 1
  else:
    return empate

```

```python
j = 11
t_mat = matrix
idx_set = range(1000)
while (len(idx_set) > 1): 
  bit = my_mode(t_mat[:,j],1)
  idx_set = [k for k in range(len(idx_set)) if t_mat[k][j]==bit]
  t_mat = t_mat[idx_set,:]
  j = j - 1

oxygen = t_mat[0]


j = 11
t_mat = matrix
idx_set = range(1000)
while (len(idx_set) > 1): 
  bit = my_antimode(t_mat[:,j],0)
  idx_set = [k for k in range(len(idx_set)) if t_mat[k][j]==bit]
  t_mat = t_mat[idx_set,:]
  j = j - 1

co2 = t_mat[0]
```
