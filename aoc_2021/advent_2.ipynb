```python
import pandas as pd
data = pd.read_table('sample_data/input_2_1.txt',header = None)
```

```python
X = 0
Y = 0

for i in range(len(data)):
  [direction, magnitude] = data[0][i].split()
  if direction == 'forward' :
    X = X + int(magnitude)
  elif direction == 'down' :
    Y = Y + int(magnitude)
  elif direction == 'up' :
    Y = Y - int(magnitude)

X*Y 
```

```python
X = 0
Y = 0
aim = 0

for i in range(len(data)):
  [direction, magnitude] = data[0][i].split()
  if direction == 'down':
    aim = aim + int(magnitude)
  elif direction == 'up':
    aim = aim - int(magnitude)
  elif direction == 'forward':
    Y = Y + aim * int(magnitude)
    X = X + int(magnitude)

X*Y 
```
