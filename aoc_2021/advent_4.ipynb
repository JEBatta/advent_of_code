```python
import pandas as pd
import numpy as np

data = pd.read_table('sample_data/input_4.txt',header = None)

```

```python
numbers = data[0][0].split(",")

n_fin = len(data)
n_tab = int((n_fin - 1)/5)
 
```

```python
bingo = []
tablero = []
for i in range(1,n_fin):
  tablero.append(data[0][i].split())
  if i % 5 == 0:
    bingo.append(np.array(tablero))
    tablero = []
```

```python
test = np.arange(0,n_tab)
test = np.delete(test,88)
test
```

```python
# Pick first winner

n_idx = 0
j = 0
win = str(np.array(['X','X','X','X','X']))
winners = []

stop_game = False
while not stop_game:

  n = numbers[n_idx]
  for j in range(n_tab):
    for i in range(5):
      if n in bingo[j][i]:
        for k in range(5):
          if n == bingo[j][i][k]:
            bingo[j][i][k]='X'
    for i in range(5):
      if win == str(bingo[j][i]) or win == str(bingo[j][:,i]):
        stop_game = True
        winners.append(j)     
  n_idx = n_idx + 1

stop_game
```

```python
# Pick last winner
tabs_in_game = np.arange(0,n_tab)
n_idx = 0
win = str(np.array(['X','X','X','X','X']))

stop_game = False
while not stop_game:
  n = numbers[n_idx]
  for j in tabs_in_game:
    for i in range(5):
      if n in bingo[j][i]:
        for k in range(5):
          if n == bingo[j][i][k]:
            bingo[j][i][k]='X'
    for i in range(5):
      if win == str(bingo[j][i]) or win == str(bingo[j][:,i]):
        tabs_in_game = np.delete(tabs_in_game,np.where(tabs_in_game == j))
        if len(tabs_in_game) == 1:
          stop_game = True    
  n_idx = n_idx + 1

stop_game
```
